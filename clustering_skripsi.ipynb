```python
def random_centroids(data, k): 
    centroids = []
    for i in range (k):
        centroid = data.apply(lambda x: float(x.sample())) 
        centroids.append(centroid)
    return pd.concat(centroids, axis=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score

# Elbow Method
sse = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(k_values, sse, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()

# Silhouette Score
silhouette_scores = []
for k in k_values[1:]: 
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(data)
    silhouette_scores.append(silhouette_score(data, labels))

plt.figure(figsize=(10, 5))
plt.plot(k_values[1:], silhouette_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Optimal k')
plt.show()

# Calculate Davies-Bouldin Index for different values of k
dbi_scores = []
k_values = range(2, 11)  
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(data)
    dbi = davies_bouldin_score(data, labels)
    dbi_scores.append(dbi)

# Plot the DBI scores
plt.figure(figsize=(10, 5))
plt.plot(k_values, dbi_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Davies-Bouldin Index')
plt.title('Davies-Bouldin Index for Optimal k')
plt.show()
```

```python
# Print SSE values for Elbow method
print("SSE values for different k:")
for k, sse_value in zip(k_values, sse):
    print(f"k={k}: SSE={sse_value}")

# Print Silhouette Scores
print("\nSilhouette Scores for different k:")
for k, silhouette_score_value in zip(k_values[0:], silhouette_scores):
    print(f"k={k}: Silhouette Score={silhouette_score_value}")

# Print Davies-Bouldin Index
print("\nDavies-Bouldin Index for different k:")
for k, dbi_value in zip(k_values, dbi_scores):
    print(f"k={k}: DBI={dbi_value}")
```

```python
def cross_validation_kmeans(data, max_clusters=10, n_splits=5):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    k_values = range(2, max_clusters + 1)
    avg_silhouette_scores = []

    for k in k_values:
        silhouette_scores = []
        for train_index, test_index in kf.split(data):
            train_data, test_data = data[train_index], data[test_index]
            kmeans = KMeans(n_clusters=k, random_state=42)
            kmeans.fit(train_data)
            labels = kmeans.predict(test_data)
            score = silhouette_score(test_data, labels)
            silhouette_scores.append(score)
        avg_silhouette_scores.append(np.mean(silhouette_scores))

    return k_values, avg_silhouette_scores
```

```python
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

k_values, avg_silhouette_scores = cross_validation_kmeans(data_scaled, max_clusters=10)

plt.plot(k_values, avg_silhouette_scores, linestyle='--', marker='o', color='b')
plt.xlabel('Number of clusters, k')
plt.ylabel('Average Silhouette Score')
plt.title('Cross-Validation Method for Optimal k')
plt.show()

optimal_k = k_values[np.argmax(avg_silhouette_scores)]
print("Optimal k:", optimal_k)
```
